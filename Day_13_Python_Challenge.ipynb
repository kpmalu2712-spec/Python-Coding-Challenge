### Project Title: Student Grades & Hobbies Tracker
### Scenario:
 A student activity coordinator is collecting information about student grades and hobbies. They want to ensure grades are updated properly and hobby lists are stored efficiently.
Tasks:
Create a dictionary with student names as keys and their grades as values.


Allow the user to add or update grades for at least 3 students.


Create a set of hobbies and allow the user to input at least 3 hobbies.


Convert the set of hobbies into a frozenset and display it.

```python
# Student Grades & Hobbies Tracker

# 1. Create a dictionary of student grades
grades = {
    "Vijay": 85,
    "Vinoth": 90,
    "Mark": 78
}

print("Initial Grades:", grades)
```

```python
# 2. Allow user to add or update grades for at least 3 students
print("\n--- Update/Add Grades for 3 Students ---")
for i in range(3):
    name = input(f"Enter student name #{i+1}: ")
    grade = int(input(f"Enter grade for {name}: "))
    grades[name] = grade   # Adds new or updates existing
    print(f"Grade updated for {name}!")

print("\nFinal Grades Dictionary:", grades)
```

```python
# 3. Create a set of hobbies and allow user to input at least 3 hobbies
hobbies = set()

print("\n--- Enter at least 3 Hobbies ---")
for i in range(3):
    hobby = input(f"Enter hobby #{i+1}: ")
    hobbies.add(hobby)

print("\nHobbies Set:", hobbies)
```

```python
# 4. Convert the set of hobbies into a frozenset and display it
hobby_frozenset = frozenset(hobbies)
print("\nHobbies Frozenset (Immutable):", hobby_frozenset)
```
