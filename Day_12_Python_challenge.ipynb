### Day 40 – List Methods and Slicing
#### Project Title: Contact Directory
### Scenario:
 You have been assigned to build a contact management system where employees can store and retrieve contact details efficiently.
Tasks:
Create a 2D list where each row holds a contact’s name, phone number, and email address.
Allow the user to add a new contact.
Display all contacts sorted alphabetically by name.
Allow the user to search for contacts by name. The search should support partial matches.
After sorting, display only the first 3 contacts in the list

```python
# 1. Create a 2D list of contacts (name, phone, email)
contacts = [
    ["Malu", "9876543210", "malu@example.com"],
    ["Nandhu",   "9123456780", "nandhu@example.com"],
    ["Priya", "9988776655", "priya@example.com"]
]
```

```python
# 2. Allow the user to add a new contact
print("Add a new contact:")
name = input("Enter name: ")
phone = input("Enter phone number: ")
email = input("Enter email: ")
contacts.append([name, phone, email])
print("\nContact added successfully!\n")
```

```python
# 3. Display all contacts sorted alphabetically by name
contacts.sort(key=lambda x: x[0])   # sort by the first element (name)

print("All Contacts (Sorted Alphabetically):")
for c in contacts:
    print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
```

```python
# 4. Allow user to search for contacts (partial matches allowed)
search = input("\nEnter name to search (partial match supported): ").lower()

print("\nSearch Results:")
found = False
for c in contacts:
    if search in c[0].lower():  # partial match check
        print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
        found = True

if not found:
    print("No matching contacts found.")
```

```python
# 5. Display only the first 3 contacts after sorting
print("\nFirst 3 contacts after sorting:")
for c in contacts[:3]:   # slicing to get first 3
    print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
```
